```python
#Calculating Score: 
score_list = []
cities = ["New York", "Los Angeles", "Chicago", "Houston", "Phoenix", "San Antonio", "Philadelphia", "San Diego", "Dallas", "Austin", "San Jose", "Fort Worth", "Jacksonville", "Charlotte", "Columbus", "Indianapolis", "San Francisco", "Seattle", "Denver", "Washington DC"]
for city in cities: 
    average = (city_dictionary[city]['aqi'] + city_dictionary[city]['pm10'] + city_dictionary[city]['pm25'] + city_dictionary[city]['o3'] + city_dictionary[city]['so2'] + city_dictionary[city]['no2'] + city_dictionary[city]['co'])/7
    score_list.append({city:average})
```
